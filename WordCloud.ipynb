## Creating Word Cloud from James Bond's Silhouette

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams['figure.figsize']=(25.0,16.0)
```

```python
script=open("bond.txt").read()
stopwords=set(STOPWORDS)
mask=np.array(Image.open("bond.jpg"))
plt.imshow(mask)
plt.axis("off")
```

```python
from matplotlib.colors import LinearSegmentedColormap as lsc
colors=["#000000","#0060A8","#484848","#FFF200"]
cmap=lsc.from_list("mycmap",colors)
wc=WordCloud(background_color="white",stopwords=stopwords,mask=mask,width=544,height=904,colormap=cmap)
wc.generate(script)
plt.figure()
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
```
